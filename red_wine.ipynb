````python
#2. importing required python libraries
import pandas as pd
import matplotlib as plt
import numpy as np
````

````python
#3. importing red_wine dataset
redwine = pd.read_csv('red_wine.csv')
````

````python
#4.previewing the first 10 rows using the head method
redwine.head(10)
````

````python
#5. previewing the last 20 rows using the tail method
redwine.tail(20)
````

````python
#6. Using the info method to check basic information on the dataset
redwine.info()
````

````python
#7. Using describe method to obtain statistical information on the datase
redwine.describe()
````

````python
#8. Histogram of numerical column "quality"
import matplotlib.pyplot as plt
plt.hist(redwine['quality'], bins=10) 
plt.xlabel('quality')
plt.ylabel('wine index')
plt.title('Quality of Wines') 
plt.show()
````

````python
#8. Scatter plot of alcohol versus quality
plt.scatter(redwine['pH'], redwine['quality'])
plt.xlabel('pH')
plt.ylabel('quality')
plt.title('pH vs. Quality') 
plt.show()
````
